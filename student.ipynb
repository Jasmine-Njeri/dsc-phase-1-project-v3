DATA EXPLORATION

Distribution of accidents over a year


```python
#Visualizing number of accidents versus year from our data

#Converting 'Event.Date' to datetime format and extract the year
df['Event.Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Event.Date'].dt.year

#Count the number of accidents per year
accidents_per_year = df['Year'].value_counts().sort_index()

#Plot the data
plt.figure(figsize=(10, 5))
sns.lineplot(x=accidents_per_year.index, y=accidents_per_year.values)
plt.title('Number of Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

From the visualization above, we can note that:

1.Before 1980, there were no aeroplanes.

2.The year 1980 recorded the highest number of accidents.

3.As the years progress, the number of accidents is reducing due to improvements in skilled manpower and technology, though more research needs to be done.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Aircraft.Category', order=df['Aircraft.Category'].value_counts().index)
plt.title('Distribution of Accidents by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.yscale('log')
plt.show()
```

```python
#Bar Plot of Accidents by Broad Phase of Flight

plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Broad.phase.of.flight', order=df['Broad.phase.of.flight'].value_counts().index)
plt.title('Number of Accidents by Broad Phase of Flight')
plt.xlabel('Number of Accidents')
plt.ylabel('Broad Phase of Flight')
plt.show()
```

Looking at the broad phase of flight:
1. Unknown - We do not know at what point the accident occured
2. The causes that we can categorize in broad phase of flight:
    1. Landing has the highest cause of accidents, followed by accidents that have happened during Takeoff.
    2. Other which has the least accidents, encompasses all the broad phase of flight not included the ones above.

```python
##Box Plot of Fatal Injuries by Aircraft Category

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Aircraft.Category', y='Total.Fatal.Injuries')
plt.title('Box Plot of Fatal Injuries by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Total Fatal Injuries')
plt.xticks(rotation=45)
plt.yscale("log")
plt.show()
```

##### The above graph shows the distribution of Fatal injuries to the make of the aircraft

```python
phase_of_flight_data = df.groupby('Broad.phase.of.flight').agg(
    total_incidents=('Event.Id', 'count'),
    total_fatal_injuries=('Total.Fatal.Injuries', 'sum'),
    total_serious_injuries=('Total.Serious.Injuries', 'sum'),
    total_minor_injuries=('Total.Minor.Injuries', 'sum')
).reset_index()
plt.figure(figsize=(14, 8))
plt.subplot(2, 1, 1)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_incidents'], color='b')
plt.title('Total Incidents by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Incidents')
plt.xticks(rotation=45)
plt.grid(True)
plt.subplot(2, 1, 2)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_fatal_injuries'], color='red', alpha=0.5, label='Fatal Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_serious_injuries'], color='orange', alpha=0.5, label='Serious Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_minor_injuries'], color='green', alpha=0.5, label='Minor Injuries')
plt.title('Total Injuries by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Injuries')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Plotting: Two subplots are created:

1. The first subplot: shows the total number of incidents by phase of flight.
2. The second subplot: shows the total number of fatal, serious, and minor injuries by phase of flight, differentiated by color.

This visualization will help you understand the distribution of incidents and injuries across different phases of flight.
